## Creating new features

**In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. Use .apply() with a custom lambda expression to create a new column called "Reason" that contains this string value.**

**Create a new column called 'Date' that contains the date from the timeStamp column. You'll need to use apply along with the .date() method.**

````python
df['Date']=df['timeStamp'].apply(lambda t: t.date())
````

**Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.**

````python
import matplotlib.pyplot as plt

# Group by Date and count calls (using any column like 'twp')
calls_per_day = df.groupby('Date').count()

# Plot the daily call counts
calls_per_day['twp'].plot(figsize=(10, 5))

# Labels
plt.xlabel('Date')
plt.ylabel('Number of 911 Calls')
plt.title('911 Calls Per Day')

plt.tight_layout()
plt.show()
````

**Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

````python
import matplotlib.pyplot as plt

# Set plot size
plt.figure(figsize=(10, 6))

# Plot each Reason separately
for reason in df['Reason'].unique():
    df_reason = df[df['Reason'] == reason]
    calls_by_date = df_reason.groupby('Date').count()
    calls_by_date['twp'].plot(label=reason)

# Labels and legend
plt.xlabel('Date')
plt.ylabel('Number of 911 Calls')
plt.title('911 Calls Per Day by Reason')
plt.legend()
plt.tight_layout()
plt.show()
````

**Now let's move on to creating  heatmaps with seaborn and our data. We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. ## Time-based Call Analysis in a Specific Township or Zip Code

### Step 1: Convert `timeStamp` to datetime

We first convert the `timeStamp` column to a proper datetime object so we can extract hour and day information.


````python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
````

### Step 2: Extract hour and day of the week

Now we extract the hour of the day and the name of the weekday from each call's timestamp.


````python
df['hour'] = df['timeStamp'].dt.hour
df['day_of_week'] = df['timeStamp'].dt.day_name()
````

### Step 3: Filter for a specific township

Let’s choose one township, for example `'NORRISTOWN'`, and filter the data to only include calls from that area.


````python
twp_filter = 'NORRISTOWN'
df_twp = df[df['twp'] == twp_filter]
````

### Step 4: Create a pivot table of call counts
We group the data by day of the week and hour, then convert it into a table format suitable for heatmapping.


````python
heatmap_data = df_twp.groupby(['day_of_week', 'hour']).size().unstack().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)
````

### Step 5: Plot a heatmap

Finally, visualize the call frequency by hour and day using a heatmap.


````python
import seaborn as sns
import matplotlib.pyplot as plt

dayHour_norristown = df_twp.groupby(by=['day_of_week', 'hour']).count()['Reason'].unstack()

plt.figure(figsize=(12, 6))
sns.heatmap(dayHour_norristown, cmap='YlOrRd', linewidths=0.5)

plt.title('Call Volume by Hour and Day in NORRISTOWN')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')

plt.tight_layout()
plt.show()
````
